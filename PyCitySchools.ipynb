```python
# Dependencies and Setup
import pandas as pd
import numpy as np
from collections import OrderedDict


# File to Load (Remember to Change These)
school_file = "Data/schools_complete.csv"
student_file = "Data/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_df = pd.read_csv(school_file)
student_df = pd.read_csv(student_file)

# Combine the data into a single dataset
complete_df = pd.merge(student_df, school_df, how="left", on=["school_name", "school_name"])

complete_df.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
total_schools = len(complete_df["school_name"].unique())
print(total_schools)

total_students = len(complete_df["student_name"])
print(total_students)

total_budget = sum(complete_df["budget"].unique())
print(total_budget)

avg_math = complete_df["math_score"].mean()
print(avg_math)

avg_reading = complete_df["reading_score"].mean()
print(avg_reading)

passing_rate = (avg_math + avg_reading)/2
print(passing_rate)

math_pass_count = complete_df.loc[complete_df["math_score"] >= 70]
math_pass_percent = ((len(math_pass_count)/total_students)*100)
print(math_pass_percent)

read_pass_count = complete_df.loc[complete_df["reading_score"] >= 70]
read_pass_percent = ((len(read_pass_count)/total_students)*100)
print(read_pass_percent)

df_data = [{"Total Number of Schools": total_schools,
            "Total Number of Students": total_students, 
            "Total District Budget": total_budget,
            "District Average Math Score": avg_math,
            "District Average Reading Score": avg_reading,
            "District Total Passing Rate": passing_rate,
            "Math Passing Rate": math_pass_percent,
            "Reading Passing Rate": read_pass_percent}]
district_df = pd.DataFrame(df_data)

district_df.head()
```

```python
summary = complete_df.groupby(["school_name"])

school_name = summary["school_name"].unique()
#print(school_name)

school_type = [x[0] for x in summary["type"].unique()]
#print(school_type)

schoolsum_stu_count = summary["student_name"].count()
#print(schoolsum_stu_count)

schoolsum_budget = summary["budget"].unique()
#print(schoolsum_budget)

school_perstu_budget = schoolsum_budget/schoolsum_stu_count
#print(school_perstu_budget)

school_avgmath = summary["math_score"].mean()
#print(school_avgmath)

school_avgread = summary["reading_score"].mean()
#print(school_avgread)

school_passmath = (complete_df[complete_df['math_score'] >= 70].groupby('school_name')['Student ID'].count()/schoolsum_stu_count) * 100
#print(school_passmath)

school_passread = (complete_df[complete_df['reading_score'] >= 70].groupby('school_name')['Student ID'].count()/schoolsum_stu_count) * 100
#print(school_passread)

school_totpass = (school_passmath + school_passread)/2
#print(school_totpass)

school_dfdata = {
                 "School Type": school_type,
                 "Total Students": schoolsum_stu_count,
                 "Total School Budget": schoolsum_budget,
                 "Per Student Budget": school_perstu_budget,
                 "Average Math Score": school_avgmath,
                 "Average Reading Score": school_avgread,
                 "Percent Passing Math": school_passmath,
                 "Percent Passing Reading": school_passread,
                 "Overall Passing Rate": school_totpass}


school_df = pd.DataFrame(school_dfdata)

school_df["Total School Budget"] = school_df["Total School Budget"].astype(int)
school_df["Per Student Budget"] = school_df["Per Student Budget"].astype(int)


school_df.head()



```

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]

spending_df = school_df[["Average Math Score",
                         "Average Reading Score",
                         "Percent Passing Math",
                         "Percent Passing Reading",
                         "Overall Passing Rate"]]

spending_df["Spending Range"] = pd.cut(school_df["Per Student Budget"], spending_bins, labels=group_names)
spending_df.head(15)
```
